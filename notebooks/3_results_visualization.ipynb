## Load data

```python
results = pd.read_csv('./output/results.txt')
results.head()
```

```python
results.columns
```

## Visualization

The following code will help you visualize the Loss and F1 score (or the metric you applied.)

```python
if ieee:
    with plt.style.context(['science', 'ieee']):
        plt.plot(results['Epoch'],results['Loss'])
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.savefig('loss.jpg', dpi=500)
        plt.show()
else:
    plt.plot(results['Epoch'],results['Loss'])
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.savefig('./output/loss.jpg')
    plt.show()
```

```python
if ieee:
    with plt.style.context(['science', 'ieee']):
        plt.plot(results['Epoch'],results['F1'])
        plt.xlabel('Epoch')
        plt.ylabel('F1 score')
        plt.savefig('f1.jpg', dpi=500)
        plt.show()
else:
    plt.plot(results['Epoch'],results['F1'])
    plt.xlabel('Epoch')
    plt.ylabel('F1 score')
    plt.savefig('./output/f1.jpg')
    plt.show()
```

```python
total_min = np.sum(results['Time(s)']) / 60
print('Time cost:', round(total_min,2), 'minutes.')
```

```python
# y-axis in log scale

plt.plot(results['Epoch'],results['Loss'])
plt.yscale('log')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
